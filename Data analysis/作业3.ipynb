## （2）数据操作

（1）读取香港酒店数据。

```python
df = pd.read_excel("香港酒店数据.xlsx")
df[:5]
```

（2）按照数据的内容，重新设置数据的索引，重新设置列名称为'名字','类型','城市','地区','地点','评分','评分人数','价格'。

```python
df.rename(columns={"字段1":'名字',"字段2":'类型',"字段3":'城市',"字段4":'地区',"字段5":'地点',"字段6":'评分',"字段7":'评分人数',"字段8":'价格'},inplace=True)
df          
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
df[df["类型"] == "浪漫情侣"][:5]
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
df[(df["类型"] == "浪漫情侣") & (df["地区"] == "湾仔")][:5]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
df[((df["地区"] == "油尖旺") & (df["评分"] > 4)) | ((df["地区"] == "观塘") & (df["评分"] > 4))][:5]
```
